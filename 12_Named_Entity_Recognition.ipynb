# Named Entity Recognition

## Aim: To implement a program that identifies the named entities in a given paragraph
## Dataset: News article


```python
import spacy
from spacy import displacy

NER = spacy.load("en_core_web_sm")
```

```python
# news Pakistan no longer can get a free terror pass for China in Afghanistan
path='/content/drive/MyDrive/NER.txt'
```

```python
f=open(path,'r')
text=f.read()
```

```python
text1= NER(text)
```

```python
for word in text1.ents:
    print(word.text,word.label_)
```

```python
displacy.render(text1,style='ent',jupyter=True)
```
